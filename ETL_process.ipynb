## 1. Извлечение данных (Extract)

```python
import pandas as pd

# Путь к вашему Excel файлу
file_path = './data/2022_синтетический_датасет.xlsx'

# Чтение данных
df = pd.read_excel(file_path)

# Вывод первых нескольких строк для оценки структуры данных
df.head()
```

## 2. Трансформация данных (Transform)

```python
# Пример трансформации: Преобразование столбца с датами
df['Дата сеанса'] = pd.to_datetime(df['Дата сеанса'], errors='coerce')
```

## 3. Загрузка данных (Load)

```python
from google.cloud import bigquery
from google.oauth2 import service_account

# Путь к файлу ключа сервисного аккаунта
key_path = 'path/to/your/service-account-file.json'

credentials = service_account.Credentials.from_service_account_file(key_path)
client = bigquery.Client(credentials=credentials, project=credentials.project_id)

# ID таблицы в формате "project_id.dataset_name.table_name"
table_id = 'your-project-id.dataset_name.table_name'

# Загрузка DataFrame в BigQuery
job = client.load_table_from_dataframe(df, table_id)

# Ожидание завершения задания
job.result()

print("Загрузка данных завершена.")
```
